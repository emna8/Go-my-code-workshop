```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Basic plotting

```python
X = [2.3, 3.6, 7.8] 
Y = [4.2 ,1.7 ,7.4]

X2 = [3.3, 4.6, 8.8] 
Y2 = [5.2 ,2.7 ,8.4]
```

## Add legend

```python
plt.plot(X,Y, linestyle='--' , color='g', label="Line 1")
plt.plot(X2,Y2, linestyle='-' , color='r', label="Line 2")
plt.legend()

plt.xlabel("X - Value")
plt.ylabel("Y - Value")

plt.title("Demo")
```
